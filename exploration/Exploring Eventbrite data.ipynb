```python
import pandas as pd
import numpy as np
from datetime import datetime

from eventbrite import Eventbrite
from pprint import pprint

eventbrite = Eventbrite('NRSZ2XEF7I2KNUIVH3FB')
cities = ['San Francisco', 'Seattle']
start_date = '2015-04-17T00:00:00Z'

def parse_event(event):
    e = {
        'city' : event['venue']['address']['city'],
        'capacity': event['capacity'],
        'start': event['start']['utc'],
        'latitude': event['venue']['address']['latitude'],
        'longitude': event['venue']['address']['longitude'],
    }
    if event.get('subcategory'):
        e['subcategory'] = event[u'subcategory'][u'name']
    if event.get('category'):
        e['category'] = event['category']['short_name']

    return e


all_events = []

for city in cities:
    page = 1

    while page < 2:
        result = eventbrite.event_search(**{'page': page,
                                            'venue.city': city,
                                            'start_date.range_start': start_date,
                                            'expand': 'category,subcategory,venue'})
        all_events += [parse_event(e) for e in result['events']]

        if page == result['pagination']['page_count']:
            break
        else:
            page += 1
    
# now we have the data in all_events. Let's dump it to a CSV
df = pd.DataFrame.from_dict(all_events)
df.to_csv('data/eventbrite_all_events_raw.csv')
```
