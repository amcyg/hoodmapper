```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

```python
# import raw crime data
sf = pd.read_csv('../downloads/sf_crime_raw.csv')

# filter on just a few categories
sf = sf[['Category', 'Location']]

# grab assault, burglary and prostitution data
sf_assault = sf[sf['Category'] == 'ASSAULT']
sf_burglary = sf[sf['Category'] == 'BURGLARY']
sf_prostitution = sf[sf['Category'] == 'PROSTITUTION']

# concatenate them together to create one crime df
crimes = [sf_assault, sf_burglary, sf_prostitution]
sf_crime = pd.concat(crimes)
sf_crime = sf_crime.reset_index()
sf_crime = sf_crime[['Category', 'Location']]

# unpack coordinates from location, fix missing negatives in longitude
sf_crime['Latitude'] = None
sf_crime['Longitude'] = None
for index in range(18770):
    coords = sf_crime['Location'][index].split(',')
    print coords
    print coords[0][1:]
    sf_crime['Latitude'][index] = float(coords[0][1:])
    sf_crime['Longitude'][index] = float(coords[1][1:-1])
    if sf_crime['Longitude'][index] > 0:
        sf_crime['Longitude'][index] = -(sf_crime['Longitude'][index])
```
