```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('../data/eventbrite_all_events_hoods.csv', header=None)
```

```python
df.head()
```

```python
df.columns = ['city', 'event_capacity', 'start_date', 'latitude', 'longitude', 'sub-category', 'category', 'hood']
```

<h3>What are the ultimate features (categories and sub-categories) that we will be comparing?</h3>

```python
print "CATEGORIES"
categories = df['category'].unique()
for category in categories:
    print category

print "SUBCATEGORIES"
subcats = df['sub-category'].unique()
for subcat in subcats:
    print subcat
```

```python
sf = df.loc[df['city'] == 'San Francisco']
```

```python
sea = df.loc[df['city'] == 'Seattle']
```

```python
for category in categories:
    sf[category] = np.nan
    sea[category] = np.nan
for subcat in subcats:
    sf[subcat] = np.nan
    sea[subcat] = np.nan
```

```python
sea_categories = sea.groupby(['hood', 'category']).count()
sea_subcats = sea.groupby(['hood', 'sub-category']).count()
```

```python
sea_categories = sea_categories['city']
sea_subcats = sea_subcats['city']
print sea_categories
```

```python
sea_categories = sea_categories.unstack(fill_value=0)
sea_subcats = sea_subcats.unstack(fill_value=0)
```

```python
sea_subcats.to_csv('../data/eventbrite_seattle_subcats.csv')
sea_categories.to_csv('../data/eventbrite_seattle_categories.csv')
```
